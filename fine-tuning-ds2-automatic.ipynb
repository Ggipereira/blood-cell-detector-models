```python
from ultralytics import YOLO
from pathlib import Path
import pandas as pd
import shutil
```

```python
PRETRAINED_MODEL = "/kaggle/input/models/guilhermegpereira/melhor-ds1-640/pytorch/default/1/best.pt"
DATA_YAML = "/kaggle/input/datasets/guilhermegpereira/ds2-3-classes/ds2 3 classes/data.yaml"
```

```python
# FASE 2: Tuning Autom√°tico
model = YOLO(PRETRAINED_MODEL)
model.tune(
    data=DATA_YAML,
    epochs=30,
    iterations=5,
    imgsz=640,
    batch=16,
    name='auto_tune',
    verbose=False
)
```
